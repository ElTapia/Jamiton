## Gaussian init condition

```python
def Q_0_gauss(x, rho_init):
    
    # Gaussiana centrada en 1500
    rho_0 = (np.exp(-((x-100)**2)/2000)/10 +rho_init) * rho_max #np.exp(-x**2/(2*2.7**2))/(2.7*np.sqrt(2*np.pi))
    
    # Menor velocidad en mayor densidad
    u_0 = 1/rho_0
    
    # y inicial en funcion de u y rho
    y_0 = rho_0 * (u_0 + h(rho_0))
    
    # Vector con codicion
    Q_0_ = np.zeros([2, len(x)])
    Q_0_[0] = rho_0
    Q_0_[1] = y_0
    
    return Q_0_
```
